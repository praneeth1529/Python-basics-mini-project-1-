```python


L1=[1,3,2,5,7]
L2=[]
def findmin(my_list):
    if not my_list:
        return None
    min_val=my_list[0]
    for item in my_list:
        if item<min_val:
            min_val=item
    return min_val

while L1:
    min_value=findmin(L1)
    L1.remove(min_value)
    L2.append(min_value)
print(f'The minvalue is {L2}')
```
